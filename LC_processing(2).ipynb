```python
spark.sql("""
create or replace view itv017907_lending_club.customers_loan_v as select
l.loan_id,
c.member_id,
c.emp_title,
c.emp_length,
c.home_ownership,
c.annual_income,
c.address_state,
c.address_zipcode,
c.address_country,
c.grade,
c.sub_grade,
c.verification_status,
c.total_high_credit_limit,
c.application_type,
c.join_annual_income,
c.verification_status_joint,
l.loan_amount,
l.funded_amount,
l.loan_term_years,
l.interest_rate,
l.monthly_installment,
l.issue_date,
l.loan_status,
l.loan_purpose,
r.total_principal_received,
r.total_interest_received,
r.total_late_fee_received,
r.last_payment_date,
r.next_payment_date,
d.delinq_2yrs,
d.delinq_amnt,
d.mths_since_last_delinq,
e.pub_rec,
e.pub_rec_bankruptcies,
e.inq_last_6mths

FROM itv017907_lending_club.customers c
LEFT JOIN itv017907_lending_club.loans l on c.member_id = l.member_id
LEFT JOIN itv017907_lending_club.loans_repayments r ON l.loan_id = r.loan_id
LEFT JOIN itv017907_lending_club.loans_defaulters_delinq d ON c.member_id = d.member_id
LEFT JOIN itv017907_lending_club.loans_defaulters_detail_rec_enq e ON c.member_id = e.member_id
""")
```

```python
spark.sql("""
create table itv017907_lending_club.customers_loan_t as select
l.loan_id,
c.member_id,
c.emp_title,
c.emp_length,
c.home_ownership,
c.annual_income,
c.address_state,
c.address_zipcode,
c.address_country,
c.grade,
c.sub_grade,
c.verification_status,
c.total_high_credit_limit,
c.application_type,
c.join_annual_income,
c.verification_status_joint,
l.loan_amount,
l.funded_amount,
l.loan_term_years,
l.interest_rate,
l.monthly_installment,
l.issue_date,
l.loan_status,
l.loan_purpose,
r.total_principal_received,
r.total_interest_received,
r.total_late_fee_received,
r.last_payment_date,
r.next_payment_date,
d.delinq_2yrs,
d.delinq_amnt,
d.mths_since_last_delinq,
e.pub_rec,
e.pub_rec_bankruptcies,
e.inq_last_6mths

FROM itv017907_lending_club.customers c
LEFT JOIN itv017907_lending_club.loans l on c.member_id = l.member_id
LEFT JOIN itv017907_lending_club.loans_repayments r ON l.loan_id = r.loan_id
LEFT JOIN itv017907_lending_club.loans_defaulters_delinq d ON c.member_id = d.member_id
LEFT JOIN itv017907_lending_club.loans_defaulters_detail_rec_enq e ON c.member_id = e.member_id
""")
```

```python
bad_data_customer_df = spark.sql("""select member_id,count(*) as total from itv017907_lending_club.customers
group by member_id having total > 1""")
```

```python
bad_data_loans_defaulters_delinq_df = spark.sql("""select member_id from(select member_id, count(*)
as total from itv017907_lending_club.loans_defaulters_delinq
group by member_id having total > 1)""")
```

```python
bad_data_loans_defaulters_delinq_df.count()
```

```python
bad_data_loans_defaulters_detail_rec_enq_df = spark.sql("""select member_id from(select member_id, count(*)
as total from itv017907_lending_club.loans_defaulters_detail_rec_enq
group by member_id having total > 1)""")
```

```python
bad_data_loans_defaulters_detail_rec_enq_df.count()
```

```python
bad_customer_data_df = bad_data_customer_df.select("member_id") \
.union(bad_data_loans_defaulters_delinq_df.select("member_id")) \
.union(bad_data_loans_defaulters_detail_rec_enq_df.select("member_id"))
```

```python
bad_customer_data_final_df =bad_customer_data_df.distinct()
```

```python
bad_customer_data_final_df.count()
```

```python
bad_customer_data_final_df.repartition(1).write \
.format("csv") \
.option("header",True) \
.mode("overwrite") \
.save("/user/itv017907/lendingclub/bad/bad_customer_data_final")
```

```python
bad_customer_data_final_df.createOrReplaceTempView("bad_data_customer")
```

```python
customers_df = spark.sql(""" SELECT * FROM 
itv017907_lending_club.customers where member_id NOT IN (select member_id from bad_data_customer)""")
```
