## Creating tables

```python
from pyspark.sql import SparkSession
import getpass
username =getpass.getuser()
spark =SparkSession.builder.config('spark.ui.port','0') \
.config('spark.shuffle.useOldFetchProtocol','true') \
.config('spark.sql.warehouse.dir', f'/user/{username}/warehouse') \
.enableHiveSupport() \
.master('yarn') \
.getOrCreate()
```

```python
#Importing for generating hash function
from pyspark.sql.functions import sha2,concat_ws
```

```python
raw_df=spark.read \
.format("csv") \
.option("inferSchema", True) \
.option("header", True) \
.load("/public/trendytech/datasets/accepted_2007_to_2018Q4.csv")
```

```python
print(f"Number of columns: {len(raw_df.columns)}")
```

```python
raw_df.count()
```

```python
raw_df.createOrReplaceTempView("raw_Data")
```

```python
spark.sql("select * from raw_Data limit 5")
```

```python
new_df = raw_df.withColumn("name_sha2",sha2(concat_ws("||",*["emp_title","emp_length","home_ownership","annual_inc","zip_code","addr_state","grade","sub_grade",
                                                    "verification_status"]),256))
```

```python
new_df.createOrReplaceTempView("newtable")
```

```python
spark.sql("select count(distinct(name_sha2)) from newtable")
```

```python
spark.sql("""select name_sha2,count(*) as total_cnt 
          from newtable group by name_sha2 having total_cnt >1 order by total_cnt DESC""")
```

```python
#Investigating outlier
spark.sql("""select * from newtable where name_sha2 like 'e3b0c44298fc1c149%'""")
```

Customers table

```python
spark.sql("""
    select name_sha2 as member_id, emp_title, emp_length,home_ownership,annual_inc,addr_state,zip_code,'USA' as country,
 grade,sub_grade,verification_status,tot_hi_cred_lim,application_type,annual_inc_joint, verification_Status_joint
 from newtable""").repartition(1).write \
.option("header",True) \
.format("csv") \
.mode("overwrite") \
.option("path",f"/user/{username}/lendingclub/raw/customers_data_csv") \
.save()
```

```python
customers_schema = 'member_id string, emp_title string, emp_length string,home_ownership string,annual_inc float,addr_State string,zip_code string,country string, grade string,sub_grade string,verification_status string,tot_hi_cred_lim float,application_type string,annual_inc_joint float,verification_status_joint string'
```

```python
customers_df = spark.read \
.format("csv") \
.option("header",True) \
.schema(customers_schema) \
.load(f"/user/{username}/lendingclub/raw/customers_data_csv")
```

Loans table

```python
spark.sql("""
    SELECT id as loan_id,name_sha2 as member_id,loan_amnt,funded_amnt,term,int_rate,installment,issue_d,loan_status,purpose,title 
    from newtable""").repartition(1).write \
.option("header",True) \
.format("csv") \
.mode("overwrite") \
.option("path",f"/user/{username}/lendingclub/raw/loans_data_csv") \
.save()
```

loan_repayments

```python
spark.sql("""
    SELECT id as loan_id,total_rec_prncp,total_rec_int,total_rec_late_fee,total_pymnt,last_pymnt_amnt,last_pymnt_d,next_pymnt_d
    from newtable""").repartition(1).write \
.option("header",True) \
.format("csv") \
.mode("overwrite") \
.option("path",f"/user/{username}/lendingclub/raw/loans_repayment_data_csv") \
.save()
```

loan_defaulters

```python
spark.sql("""
    SELECT name_sha2 as member_id,delinq_2yrs,delinq_amnt,pub_rec,pub_rec_bankruptcies,
    inq_last_6mths,total_rec_late_fee, mths_since_last_delinq, mths_since_last_record
    from newtable""").repartition(1).write \
.option("header",True) \
.format("csv") \
.mode("overwrite") \
.option("path",f"/user/{username}/lendingclub/raw/loans_defaulter_csv") \
.save()
```
